## 2.3.1 Joint Probablity Distributions

* **Joint Probability**: The probablity of two or more events happening together
* **Joint Probability Distribution**: Shows a probabilty distribution for two or more variables.
* **Two Discrete Random Variables**:
  * Let $X$ and $Y$ be two discrete random variables define by the sample space S. The joint probablity mass function, p(x,y) is defined for each pair by
    * $p(x,y)$ = $P(X = x$ and $Y = y)$
      * Also, it needs to satisfy $p(x,y) ≥ 0$ and $∑_{x,y} p(x,y) = 1$
  * **Marginal Distribution**: The marginal distribution of a subset is the probablity of each variable in random variable $X$ and $Y$, without reference for each other.
    * The marginal probability mass function of X is defined as
        * $P_{X}(X) = ∑_{y:p(x,y)>0} p(x,y)$ for all x
    * Similiarly defined for Y,
        * $P_{Y}(Y) = ∑_{x:p(x,y)>0} p(x,y)$ for all y

* **Two Continuous Random Variables**
  * Let $X$ and $Y$ be continuous random variables. The joint pdf, f(x,y), is (for any two-dimensional set A)
    * $P[(X,Y) ∈ A] = ∫∫_{A}f(x,y)dxdy$
      * Also, it needs to satisfy $f(x,y) ≥ 0$ and $∫^{B}$ $∫^{B} f(x,y)dxdy = 1$, where B is $∫^{∞}_{-∞}$ (It's glitch so I can't do double integrals of ∞ to -∞)
    * **Marginal Distribution**: Similiarly like the two discrete variable, but woth some differences.

* **Independent Random Variables**: Two random variables $X$ and $Y$ are said to be independent, if for every pair x and y
  * $p(x,y) = p_{X}(x) · p_{Y}(y)$ for discrete
  * $f(x,y) = f_{X}(x) · f_{Y}(y)$ for continuous

```python
import random
#Example of discrete random variable

#Number of trials
trials = 1000
#Number of successes for Joint Probability
X_Y = 0

for i in range(trials):
  die_1 = random.randint(1,6)
  die_2 = random.randint(1,6)

  #Let event X be the event that die 1 rolls a 4
  X = (die_1 == 4)
  #Let event Y be the event that die 2 rolls an even number
  Y = (die_2 % 2 == 0)

  if X and Y:
    X_Y += 1
#The joint probability is the number of times both X and Y happened divided by the number of trials
joint_prob = X_Y/trials

print("Joint Probalilty P(X and Y): ", joint_prob)

# THe probabilty of X and Y multiplied
exp_joint_prob = (1/6) * (1/2)
print("Expected probability: ", exp_joint_prob)
```

Joint probability is the probability of two events. In this scenario, it is the event of rolling a 4 with die one and rolling an even number with die two. The joint probability is an real random example, while the expected probabilty is the theoretical probabilty of the X and Y. The real random example should be relatively near the theoretical probability.

## 2.3.2 Correlation and Dependence

* **Correlation**: The relation between two variables
* **Covariance**: Covariance is the measure of the joint variability of two variables. Let $X$ and $Y$ be jointly distributed random variables with pmf $p(x,y)$ or pdf $f(x,y)$, discrete and continuous respectively. The equation is
  * $Cov(X,Y) = E[(X-μ_{X})(Y-μ_{Y})]$
      * Use $∑$ for discrete and $∫$ for continuous
  * $Cov(X,X) = E[(X - \mu_{X})^{2}] = V(X)$
  * **Correlation Coefficient**: The covariance of two random variables divided by theproduct of their standard deviation, which measures the linear correlation between two variables.
    * $ρ_{X,Y}$ = $COV(X,Y) / σ · σ$
      * The two standard deviations above belong to X and Y respectively
    * If $X$ and $Y$ are independent, then $ρ = 0$. However, $ρ = 0$ doesn't mean independence

## 2.3.3 Random samples

* **Simple Random Sample**: A randomly selected subset of a population.
  * Random variables $X_{1}, X_{2},... X_{n}$ form a simple random sample of size n if
    * $X_{i}'s$ are independent from each other
    * Every $X_{i}$ has the same probablity distribution

* **Central Limit Theorem**: The properly normalized sum of independent random variables tend to a normal sitribution even if the original variables($X_{1}, X_{2}...$) are not normally distributed.
  * Let $X_{1}, X_{2}... X_{n}$ be a random sample from a distribution with mean $μ$ and varience $σ^{2}$, then if n is sufficiently large, then $\hat{X}$ has a normal distribution with $μ_{̂\hat{X}} = μ$ and $σ^{2}_{\hat{X}}/n$.
