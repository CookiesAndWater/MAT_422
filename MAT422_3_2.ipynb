# 3.2.1 Limits and Continuity

* **Euclidean norm**: $||x|| = \sqrt{∑^D_{i=1}x^2_i}$, for all x = ()
* **Open r-ball**: The open r-ball around $x \in \mathbb{R}^d$ is the set of points within Euclidean distance r of x, such that
  * $B_r(x) = \{y ∈ \mathbb{R}^d : ||y-x|| < r \}$
* **Limit Point (Accumulation Point)**: A point $x ⊆ \mathbb{R}^d$ is a limit point of a set $A ⊆ \mathbb{R}^d$ if every open ball around x contains a point from $a$, where $a \neq x$.
* **Sets**:
  * **Closed Set**: A set $A$ is closed if every limit point A belongs to A.
  * **Opened Set**: A set $A$ is open if for every $x \in A$, $B_r(x) ⊆ A$
  * **Bounded Set**: A set $A$ is bounded if there exist an $r>0$ such that $A ⊆ B_r(0),$ where $0 = (0,...,0)^T$
* **Limits of a function**: Let $f: D → \mathbb{R}^d$. A function $f$ has a limit $L$ as $x →a$, if
  * For any $ϵ > 0$, there exists a $δ > 0$ that $|f(x) - L| < ϵ$, for $x ∈ D ∩ B_δ(a)$, not including the set ${a}$.
    * Or simply lim$_{x→a}f(x)=L$
* **Continuous Function**: A function that doesn't have any discontinuities.  
  * Let $f: D → \mathbb{R}$ on $D ⊆ \mathbb{R}^d$. Then f is continuous at $a ∈ D$ if
    * lim$_{x→a}f(x)=f(a)$
* **Composition of functions**: If $f$ is continuous at $x_0$, and $g$ in continuous at $f(x_0)$, the the composition $g ∘ f$ is continuous at $x_0$
* **Extrema**: Let $f: D → \mathbb{R}$. A function $f$ attains a maximum, $M$, at $z$* if $f(z$*$) = M$ and $M ≥ f(x)$ for all $x ∈D$. Similiarly, $f$ attains a minimum, m, at $z_*$ if $f(z_*) = m$ and $m ≤ f(x) for all x ∈ D$
* **Extreme Value Thereom**: A continuous function on a nonempty, closed, bounded set will attain both a maximum and a minimum value.

# 3.2.2 Derivatives

## 3.2.2.1 Single-Variable Case

* **Derivative Definition**: For a function $f:D → \mathbb{R}$, and where $x_0 \in D$ and is an interior point, then the derivative of $f$ at $x_0$ is
  * $f'(x_0) = \frac{df(x_0)}{dx} =  lim_{h → 0} \frac{f(x_0+h) - f(x_0)}{h}$, if the limit exists.
* **Linearity of Derivatives**: If $f$ and $g$ are differentiable at $x$, and $α$ and $β$ are constants, then
  * $[αf(x) + βg(x)]' = αf'(x) + βg'(x)$
* **Extreme Values Insight**:
  * If $f'(x_0) > 0, f(x) > f(x_0)$ for $x > x_0$ and $f(x) < f(x_0)$ for $x < x_0$
  * If $f'(x_0) < 0$, then the opposite above(switch the inequalities).
* **Rolle's Theorem**: Let $f:[a,b] → \mathbb{R}$ be a continuous function with $f(a) = f(b)$, there exists $c ∈ (a,b)$ such that
  * $f'(c) = 0$
* **Mean Value Thereom**: Let $f:[a,b]→ \mathbb{R}$ be a continuous function and its derivatives exist on (a,b). Then there is $a < c < b$ such that
  * $f(b) = f(a) + (b-a)f'(c)$
    * In terms of $f'(c)$, $f'(c) = \frac{f(b) - f(a)}{b-a}$

## 3.2.2.2 General Case

* **Partial derivative**: For a function $f:D → \mathbb{R}$, and where $x_0 \in D$ and is an interior point, then the partial derivative of $f$ at $x_0$  with respect to $x_i$ is
  * $\frac{∂f(x_0)}{∂x_i} = lim_{h→0}\frac{f(x_0 + he_i) - f(x_0)}{h}$
* **Jacobian**: For a vector-valued function $f = (f_1,...,f_m): D → \mathbb{R}^m$, the jacobian is the matrix of all first-order partial derivatives.
* **Gradient**: For a real-valued function $f$, the gradient $\nabla f(x_0)$ is a vector of partial derivatives.
* **Chain Rule**: For differentiable functions $f$ and $g$:
  * $J_{g∘f}(x_0) = J_g(f(x_0))J_f(x_0)$ as product of matrices

## 3.2.2.3 Further Derivatives

* **Directional Derivative**: The directional derivative of $f$ at $x_0$ in the direction of a unit vector v is
  * $\frac{∂f(x_0)}{∂v} = lim_{h→0}\frac{f(x_0 + hv) - f(x_0)}{h}$
* **Directional Derivative from Gradient**: If $f$ is continuously differentiable at $x_0$, then
  * $\frac{∂f(x_0)}{∂v} = \nabla f(x_0)^Tv$

# 3.2.3  Taylor's Theorem

* **Taylor Thereom**: Let $f : D → \mathbb{R}$, where $D ⊆ \mathbb{R}$. Also suppose f has m times continuous derivatives on [a,b]. Then
  * $f(b) = f(a) + (b-a)f'(a) + \frac{1}{2}(b-a)^2f''(a) + \frac{(b-a)^{m-1}}{(m-1)!}f^{m-1}(a) + R_m$
    * Where $R_m = \frac{(b-a)^m}{m!}f^m(a+θ(b-a))$ for some $0 < θ < 1$.
  * What the Taylor series does is approximate a function using it's derivatives, as the derivatives of a function define how it's changing.

```python
# Taylor series example for e^x
import numpy as np
import matplotlib.pyplot as plt

x_val = np.linspace(-2,2,400)
actual_val = np.exp(x_val)

def taylor_series(x,n):
    return sum((x**k) / np.math.factorial(k) for k in range(n))

approx = [taylor_series(x,5) for x in x_val]

plt.figure(figsize=(10,6))
plt.plot(x_val, actual_val, label = "e^x", color = "blue")
plt.plot(x_val, approx, label = "Taylor Series Approximation(n=5)", color = "red")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

The taylor series for $e^x$ is $∑^∞_{n=0}\frac{x^n}{n!}$. So in the taylor series method above, we write that equation. We use n = 5, which means that we approximate that to the fifth derivative. In the graph above, we can see the actual function $e^x$ and it's approximation. Even with only the fifth derivative, the approximation is close to the original.
